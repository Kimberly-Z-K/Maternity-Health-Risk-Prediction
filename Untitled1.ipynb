```python
import pandas as pd
import numpy as np
import seaborn as sns
import sklearn as sk
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('/content/Maternal Health Risk Data Set.csv');

nulls=pd.isnull(data).sum()
dtype =data.dtypes
dhead=data;
print(dhead)
#dups=data.apply(lambda x: sum(x.duplicated()))
#duploc=data.duplicated().sum()
data=data.drop_duplicates()
duploc=data.duplicated().sum()
#print(duploc)


```

```python
X=data[['Age','SystolicBP','DiastolicBP','BS','BodyTemp','HeartRate']]
y=data['RiskLevel']
```

```python
X_train ,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.20,random_state=45)
```

```python
forest=RandomForestClassifier(random_state=45)


gradBoost=GradientBoostingClassifier(random_state=45)
Stacker=StackingClassifier(estimators=[('rf',forest),('gb',gradBoost)],final_estimator=LogisticRegression())
Stacker.fit(X_train,Y_train)
```

```python
prediction=Stacker.predict(X_test)
```

```python
accuracy=accuracy_score(prediction,Y_test)
print(f'Accuracy is:    {accuracy}')
confusion=confusion_matrix(prediction,Y_test)

plt.figure(figsize=(6,5))
sns.heatmap(confusion,annot=True,fmt='d',cmap='BuGn',xticklabels=['Low Risk','medium Risk','High risk'],yticklabels=['Low Risk','Medium Risk','High Risk'])

plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.title('Accuracy Confusion metrics')
plt.show()
```
